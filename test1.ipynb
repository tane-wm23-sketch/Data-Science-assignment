## Data preprocessing

### Subtask:
Handle any missing values, outliers, or inconsistencies in the data.


**Reasoning**:
Check for missing values in X and y, identify potential outliers in X using IQR, and decide on an outlier handling strategy.



```python
print("Missing values in X:")
print(X.isnull().sum().sum())
print("\nMissing values in y:")
print(y.isnull().sum().sum())

# Identify outliers using IQR
Q1 = X.quantile(0.25)
Q3 = X.quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = ((X < lower_bound) | (X > upper_bound)).sum()

print("\nNumber of outliers per feature (using IQR):")
print(outliers[outliers > 0])

# Strategy: Capping outliers to the bounds
X_capped = X.copy()
for column in X_capped.columns:
    X_capped[column] = X_capped[column].clip(lower=lower_bound[column], upper=upper_bound[column])

X = X_capped
display(X.head())
```

**Reasoning**:
Given the severe class imbalance identified in the previous step, apply an appropriate resampling technique (SMOTE) to balance the dataset.



```python
from imblearn.over_sampling import SMOTE
import pandas as pd

# Apply SMOTE to the training data to handle class imbalance
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

print("Shape of X after resampling:", X_resampled.shape)
print("Shape of y after resampling:", y_resampled.shape)
print("\nValue counts of y after resampling:")
print(y_resampled.value_counts())
```

## Feature selection/engineering

### Subtask:
Select relevant features or create new ones that could improve the model's performance.


**Reasoning**:
Import necessary libraries, instantiate SelectKBest, fit it to the resampled data, get scores and p-values, create a results DataFrame, filter based on p-value, select features, and display the shape of the original and selected dataframes.



```python
from sklearn.feature_selection import SelectKBest, chi2
import pandas as pd
import seaborn as sns
# Instantiate SelectKBest with chi2 score function and k='all'
selector = SelectKBest(score_func=chi2, k='all')

# Fit SelectKBest on the resampled data
selector.fit(X_resampled, y_resampled.values.ravel())

# Get the feature scores and p-values
feature_scores = pd.Series(selector.scores_, index=X_resampled.columns)
feature_pvalues = pd.Series(selector.pvalues_, index=X_resampled.columns)

# Create a DataFrame to store the feature names, scores, and p-values
feature_selection_results = pd.DataFrame({
    'Feature': X_resampled.columns,
    'Score': feature_scores,
    'P_Value': feature_pvalues
})

# Filter features based on a significance level (e.g., p < 0.05)
selected_features_df = feature_selection_results[feature_selection_results['P_Value'] < 0.05]
selected_features = selected_features_df['Feature'].tolist()

# Create a new DataFrame X_selected containing only the selected features
X_selected = X_resampled[selected_features]

# Display the shape of the original X_resampled and the selected X_selected DataFrames
print("Shape of original X_resampled:", X_resampled.shape)
print("Shape of selected X_selected:", X_selected.shape)

display(selected_features_df.sort_values(by='Score', ascending=False))

feature_selection_results_sorted = feature_selection_results.sort_values(by='Score', ascending=False)

```
